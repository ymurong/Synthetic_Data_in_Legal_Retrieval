```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_tsdae = pd.read_excel('tsdae_zeroshot.xlsx', index_col=None)
df_tsdae
```

```python
x = df_tsdae['steps']
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size

for column in df_tsdae.columns[1:]:  # Skip the first column for the x-axis
    plt.plot(x, df_tsdae[column], label=column)

# Adding titles and labels
plt.title('TSDAE Zeroshot Performance')
plt.xlabel('Steps')
plt.ylabel('Retrieval Performance(in percent)')
plt.legend()
```
