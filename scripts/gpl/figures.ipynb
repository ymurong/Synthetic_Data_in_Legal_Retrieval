```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_gpl = pd.read_excel('gpl_zeroshot.xlsx', index_col=None)
df_gpl
```

```python
x = df_gpl['steps']
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size

for column in df_gpl.columns[1:]:  # Skip the first column for the x-axis
    plt.plot(x, df_gpl[column], label=column)

# Adding titles and labels
plt.title('GPL Zeroshot Performance')
plt.xlabel('Steps')
plt.ylabel('Retrieval Performance(in percent)')
plt.legend()
```
